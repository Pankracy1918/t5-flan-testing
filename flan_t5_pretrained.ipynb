```python
model = AutoModelForSeq2SeqLM.from_pretrained('google/flan-t5-large')

tokenizer = AutoTokenizer.from_pretrained('google/flan-t5-large')

```

```python
print(model.config)
```

```python

%%time 
task = "Math reasoning" #@param ["Translate", "Question Answering", "Logical reasoning", "Scientific knowledge", "Yes/no question", "Reasoning task", "Boolean Expressions", "Math reasoning", "Premise and hypothesis"]

text = "Why Huggy Wuggy eats people?" #@param {type: 'string'}

input_text = task + text

if task == "Translate":
  input_text = "Translate to " + text
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids)
 

if task == "Question Answering":
  input_text = "Please answer to the following question. " + text
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids, \
                         min_length=256, \
                         max_new_tokens=256, \
                         length_penalty = 2, \
                         num_beams=16, \
                         no_repeat_ngram_size=2, \
                         #num_return_sequences =2 \
                         early_stopping=True)


if task == "Logical reasoning":
  input_text = "Q: " + text + ".Give the rationale before answering."
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids, \
                         min_length=128, \
                         max_new_tokens=128, \
                         length_penalty = 2, \
                         num_beams=16, \
                         no_repeat_ngram_size=2, \
                         #num_return_sequences =2 \
                         early_stopping=True)


if task == "Scientific knowledge":
  input_text = "Please answer the following question. " + text
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids, \
                         min_length=128, \
                         max_new_tokens=128, \
                         length_penalty = 2, \
                         num_beams=16, \
                         no_repeat_ngram_size=2, \
                         #num_return_sequences =2 \
                         early_stopping=True)


if task == "Yes/no question":
  input_text = "Answer the following yes/no question." + text
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids)

if task == "Reasoning task":
  input_text = "Answer the following yes/no question by reasoning step-by-step. " + text
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids, \
                         min_length=64, \
                         max_new_tokens=64, \
                         length_penalty = 2, \
                         num_beams=16, \
                         no_repeat_ngram_size=2, \
                         #num_return_sequences =2 \
                         early_stopping=True)


if task == "Boolean Expressions":
  input_text = "Q: ("+text+") is? A: Let's think step by step"
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids)


if task == "Math reasoning":
  input_text = text
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids)


if task == "Premise and hypothesis":
  input_text = text
  input_ids = tokenizer(input_text, return_tensors='pt').input_ids
  outputs = model.generate(input_ids, \
                         min_length=256, \
                         max_new_tokens=256, \
                         length_penalty = 2, \
                         num_beams=16, \
                         no_repeat_ngram_size=2, \
                         #num_return_sequences =2 \
                         early_stopping=True)



print(input_text)
print(tokenizer.decode(outputs[0],skip_special_tokens=True))
```
